```python

# Subtract each element (excluding the 'date' column) from 1
air_quality_index = sum_delta_divided.apply(lambda x: 1 - x)

air_quality_index['date'] = date_range

# Set 'date' column as the index
air_quality_index.set_index('date', inplace=True)

# Display the result
air_quality_index
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
from mpl_toolkits.axes_grid1 import make_axes_locatable
import os

# Transpose the DataFrame
transposed_df = air_quality_index.transpose()

# Extract latitude and longitude from the index
transposed_df[['Latitude', 'Longitude']] = transposed_df.index.to_series().apply(lambda x: pd.Series(eval(x)))

# Create a GeoDataFrame with Point geometries based on latitude and longitude
gdf = gpd.GeoDataFrame(transposed_df, geometry=gpd.points_from_xy(transposed_df['Longitude'], transposed_df['Latitude']))

# Set the path to the Downloads folder
downloads_folder = './Karnataka_shape_file'  # Change folder name accordingly
pic_save_folder = "./contour_plot_karnataka"

# Load the Karnataka shapefile
kar_map = gpd.read_file(os.path.join(downloads_folder, 'Karnataka.shp'))  # Replace with the actual path to the shapefile

# Check if points are within the Karnataka boundary
points_within_kar = gdf[gdf.geometry.within(kar_map.unary_union)]

# Loop through each day and create a plot for points within Karnataka
for date_column in transposed_df.columns[:-2]:
    # Set up the plot
    fig, ax = plt.subplots(figsize=(10, 8))
    
    # Plot the base map of Karnataka
    kar_map.plot(ax=ax, color='white', edgecolor='black')

    # Scatter plot with color representing the air quality index values for points within Karnataka
    if not points_within_kar.empty:
        scatter = ax.scatter(points_within_kar['Longitude'], points_within_kar['Latitude'],
                             c=points_within_kar[date_column], cmap='viridis', s=50, edgecolors='black')

        # Create colorbar
        divider = make_axes_locatable(ax)
        cax = divider.append_axes("right", size="5%", pad=0.1)
        plt.colorbar(scatter, cax=cax, label='Air Quality Index')

    ax.set_title(f'Air Quality Index Contour Plot - {date_column}')
    
    # Save the figure in the Downloads folder with a unique filename
    fig.savefig(os.path.join(pic_save_folder, f'air_quality_index_Plot_{date_column}.png'))
    
    # Close the figure to free up resources
    plt.close()
```
