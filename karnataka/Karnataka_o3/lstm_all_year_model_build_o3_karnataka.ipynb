```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Conv1D, MaxPooling1D
from tensorflow.keras.layers import Conv1D, AveragePooling1D

data = np.array(numeric_df)

# Create sequences of data with a window size of 5
sequence_length = 5
sequences = []
next_day_values = []

for i in range(len(data) - sequence_length):
    sequences.append(data[i:i+sequence_length])
    next_day_values.append(data[i+sequence_length])

# Convert sequences and next_day_values to NumPy arrays
sequences = np.array(sequences)
next_day_values = np.array(next_day_values)

# Split the data into training and test sets
#X_train, X_test, y_train, y_test = train_test_split(sequences, next_day_values, test_size=0.2, random_state=42)

# Define and compile the CNN-LSTM hybrid model
model_3 = Sequential()
model_3.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(sequence_length, data.shape[1])))
model_3.add(AveragePooling1D(pool_size=3))
model_3.add(LSTM(64, activation='relu'))  # Add LSTM layer with 64 units and ReLU activation
model_3.add(Dense(data.shape[1], activation='relu'))  # Output layer with ReLU activation

model_3.compile(loss='mean_squared_error', optimizer='adam')

# Fit the model on the training data
model_3.fit(sequences, next_day_values, epochs=500, batch_size=32, verbose=1)

# Make predictions on the test set
test_predictions = model_3.predict(sequences)

# Apply ReLU activation to enforce non-negativity in predictions
test_predictions = np.maximum(test_predictions, 0)


gh = data[sequence_length:]
# Find the index of the rows where the full row does not contain zero
non_zero_indices = np.where(~np.all(next_day_values == 0, axis=1))[0]

# Obtain the array that does not contain any zeros
non_zero_array = next_day_values[non_zero_indices]
testing = test_predictions[non_zero_indices]

import numpy as np
from sklearn.metrics import mean_squared_error

# Assuming your data is stored in the variables 'testing_array' and 'non_zero_array'
mse_list_3 = []

# Iterate through each row (day) in the arrays
for i in range(testing.shape[0]):
    # Calculate the mean squared error for the current day
    mse = mean_squared_error(testing[i], non_zero_array[i])
    mse_list_3.append(mse)

# Convert the list of MSE values to a NumPy array
mse_array_3 = np.array(mse_list_3)

# Print the mean squared errors for each day
for i, mse in enumerate(mse_array_3,5):
    print(f"Day {i + 1}: MSE = {mse}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
from mpl_toolkits.axes_grid1 import make_axes_locatable
import os

# Assuming 'result_df' is your dataframe with original and predicted values

# Create a date range from "6/9/2023" to "30/9/2023" and exclude 30/9/23
date_range = pd.date_range(start="2023-09-06", end="2023-09-30")

# Exclude the day '2023-09-30'
date_range_excluded = date_range[date_range != '2023-09-30']

# Display the resulting DatetimeIndex
date_range = date_range_excluded

# Set the date range as the index of the dataframe
predicted_df.index = date_range

# Transpose the DataFrame
transposed_o3_df = predicted_df.transpose()

# Extract latitude and longitude from the index
transposed_o3_df['Latitude'] = transposed_o3_df.index.map(lambda x: x[0])
transposed_o3_df['Longitude'] = transposed_o3_df.index.map(lambda x: x[1])

# Create a GeoDataFrame with Point geometries based on latitude and longitude
gdf = gpd.GeoDataFrame(transposed_o3_df, geometry=gpd.points_from_xy(transposed_o3_df['Longitude'], transposed_o3_df['Latitude']))

# Set the path to the Downloads folder
downloads_folder = '/home/sysadm/Downloads/karnataka/Karnataka_o3/Karnataka_shape_file'  # Change folder name accordingly
pic_save_folder = "/home/sysadm/Downloads/karnataka/Karnataka_o3/contour_plot"


# Load the Karnataka shapefile
karnataka_map = gpd.read_file(os.path.join(downloads_folder, 'Karnataka.shp')) 
#Check if points are within the Karnataka boundary
points_within_karnataka = gdf[gdf.geometry.within(karnataka_map.unary_union)]

# Loop through each day and create a plot for points within Karnataka
for date_column in transposed_o3_df.columns[:-2]:
    # Set up the plot
    fig, ax = plt.subplots(figsize=(10, 8))
    
    # Plot the base map of Karnataka
    karnataka_map.plot(ax=ax, color='white', edgecolor='black')

    # Scatter plot with color representing the O3 values for points within Karnataka
    if not points_within_karnataka.empty:
        scatter = ax.scatter(points_within_karnataka['Longitude'], points_within_karnataka['Latitude'],
                             c=points_within_karnataka[date_column], cmap='viridis', s=50, edgecolors='black')

        # Create colorbar
        divider = make_axes_locatable(ax)
        cax = divider.append_axes("right", size="5%", pad=0.1)
        plt.colorbar(scatter, cax=cax, label='O3 Value')

    ax.set_title(f'O3 Contour Plot - {date_column}')
    
    # Save the figure in the Downloads folder with a unique filename
    fig.savefig(os.path.join(pic_save_folder, f'o3_Plot_{date_column}.png'))
    
    # Close the figure to free up resources
    plt.close()
```
