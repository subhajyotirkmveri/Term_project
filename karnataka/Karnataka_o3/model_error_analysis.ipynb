```python
import matplotlib.pyplot as plt
import numpy as np

# Load kriging data for O3 in Karnataka
krig_karnataka_o3 = pd.read_csv("/home/sysadm/Downloads/karnataka/Karnataka_2023_sep_O3_final.csv")
kriging_remove_karnataka_o3 = krig_karnataka_o3.drop(columns=["latitude", "longitude", "2023-09-01", "2023-09-02", "2023-09-03", "2023-09-04", "2023-09-05"])

# Load LSTM predicted data for O3 in Karnataka
lstm_karnataka_o3 = pd.read_csv("/home/sysadm/Downloads/karnataka/Karnataka_o3/relu_predicted_6_30_o3.csv")
lstm_karnataka_o3.index = lstm_karnataka_o3["date"]
lstm_karnataka_o3 = lstm_karnataka_o3.drop(columns=['date'])

# Transpose LSTM predictions for O3 in Karnataka
relu_predicted_karnataka_o3 = lstm_karnataka_o3.transpose()
df_karnataka_o3 = relu_predicted_karnataka_o3
kriging_remove_karnataka_o3.index = relu_predicted_karnataka_o3.index
df1_karnataka_o3 = kriging_remove_karnataka_o3.to_numpy()
df2_karnataka_o3 = relu_predicted_karnataka_o3.to_numpy()
result_df_numpy_karnataka_o3 = df1_karnataka_o3 - df2_karnataka_o3

# Calculate the absolute value of each element
abs_array_karnataka_o3 = np.abs(result_df_numpy_karnataka_o3)

# Print or use the result array as needed
print(abs_array_karnataka_o3)

abs_array_df_karnataka_o3 = pd.DataFrame(abs_array_karnataka_o3, columns=relu_predicted_karnataka_o3.columns)
abs_array_df_karnataka_o3["longitude"] = krig_karnataka_o3["longitude"]
abs_array_df_karnataka_o3["latitude"] = krig_karnataka_o3["latitude"]
result_df_karnataka_o3 = abs_array_df_karnataka_o3

import geopandas as gpd
import matplotlib.pyplot as plt

df2_karnataka_o3 = result_df_karnataka_o3

# Assuming your DataFrame is named 'your_dataframe'
# Replace 'your_dataframe' with your actual DataFrame name
gdf_karnataka_o3 = gpd.GeoDataFrame(df2_karnataka_o3, geometry=gpd.points_from_xy(df2_karnataka_o3['longitude'], df2_karnataka_o3['latitude']))

# Replace 'your_shapefile.shp' with your actual file path for Karnataka
shapefile_path_karnataka_o3 = './Karnataka_shape_file/Karnataka.shp'
basemap_karnataka_o3 = gpd.read_file(shapefile_path_karnataka_o3)

# Check if points are within the shapefile boundary
points_within_shapefile_karnataka_o3 = gdf_karnataka_o3[gdf_karnataka_o3.geometry.within(basemap_karnataka_o3.unary_union)]
gdf_within_shapefile_karnataka_o3 = points_within_shapefile_karnataka_o3.reset_index(drop=True)  # Reset the index to avoid potential issues

# Now you can use gdf_within_shapefile_karnataka_o3 for further analysis or visualization

result_df_karnataka_o3 = gdf_within_shapefile_karnataka_o3

# Iterate through each date column in the DataFrame
for column in result_df_karnataka_o3.columns[0:-3]:  # Exclude latitude and longitude columns
    # Check if the entire column has NaN values
    if not result_df_karnataka_o3[column].isnull(). all():
        # Extract only the date part from the datetime string
        date_only_karnataka_o3 = pd.to_datetime(column).date()

        # Pivot the DataFrame for the specific date
        heatmap_data_karnataka_o3 = result_df_karnataka_o3.pivot(index='latitude', columns='longitude', values=column)

        # Create a contour plot if there are no NaN values for the specific date
        if not heatmap_data_karnataka_o3.isnull().values.all():
            plt.figure(figsize=(10, 6))
            contour_karnataka_o3 = plt.contourf(heatmap_data_karnataka_o3.columns, heatmap_data_karnataka_o3.index, heatmap_data_karnataka_o3.values, cmap='winter')
            plt.colorbar(contour_karnataka_o3, label='Error O3 Value')
            plt.xticks(heatmap_data_karnataka_o3.columns[::5])
            plt.yticks(heatmap_data_karnataka_o3.index[::5])
            plt.xlabel('Longitude')
            plt.ylabel('Latitude')
            plt.title(f'Error Analysis for O3 Levels in Karnataka - {date_only_karnataka_o3}')
            plt.grid()
            plt.savefig(f"/home/sysadm/Downloads/karnataka/Karnataka_o3/error_plot/Error Analysis for O3_plot_heat_{date_only_karnataka_o3}_karnataka.png")
            plt.show()
```
