```python
import matplotlib.pyplot as plt
import numpy as np

krig_o3 = pd.read_csv("Delhi_2023_sep_O3_final.csv")
krig_o3
kriging_remove = krig_o3.drop(columns = ["latitude","longitude","2023-09-01","2023-09-02","2023-09-03","2023-09-04","2023-09-05"])

lstm_o3 = pd.read_csv("/home/sysadm/Downloads/delhi/delhi_o3/relu_predicted_6_30_o3.csv")
lstm_o3.index = lstm_o3["date"]
lstm_o3 = lstm_o3.drop(columns=['date'])

relu_predicted = lstm_o3.transpose()
df = relu_predicted
kriging_remove.index = relu_predicted.index
df1 = kriging_remove.to_numpy()
df2 = relu_predicted.to_numpy()
result_df_numpy = df1-df2

# Calculate the absolute value of each element
abs_array = np.abs(result_df_numpy)

# Print or use the result array as needed
print(abs_array)

abs_array_df = pd.DataFrame(abs_array,columns = relu_predicted.columns)
abs_array_df["longitude"] = krig_o3["longitude"]
abs_array_df["latitude"] = krig_o3["latitude"]
result_df = abs_array_df
import geopandas as gpd
import matplotlib.pyplot as plt
df2 = result_df
# Assuming your DataFrame is named 'your_dataframe'
# Replace 'your_dataframe' with your actual DataFrame name
gdf = gpd.GeoDataFrame(df2, geometry=gpd.points_from_xy(df2['longitude'], df2['latitude']))

# Replace 'your_shapefile.shp' with your actual file path
shapefile_path = './Delhi_shape_file/Delhi.shp'
basemap = gpd.read_file(shapefile_path)

# Check if points are within the shapefile boundary
points_within_shapefile = gdf[gdf.geometry.within(basemap.unary_union)]
gdf_within_shapefile = points_within_shapefile.reset_index(drop=True)  # Reset the index to avoid potential issues
gdf_within_shapefile
# Now you can use gdf_within_shapefile for further analysis or visualization

result_df= gdf_within_shapefile

# Iterate through each date column in the DataFrame
for column in result_df.columns[0:-3]:  # Exclude latitude and longitude columns
    # Check if the entire column has NaN values
    if not result_df[column].isnull().all():
        # Extract only the date part from the datetime string
        date_only = pd.to_datetime(column).date()
        
        # Pivot the DataFrame for the specific date
        heatmap_data = result_df.pivot(index='latitude', columns='longitude', values=column)
        
        # Create a contour plot if there are no NaN values for the specific date
        if not heatmap_data.isnull().values.all():
            plt.figure(figsize=(10, 6))
            contour = plt.contourf(heatmap_data.columns, heatmap_data.index, heatmap_data.values, cmap='winter')
            plt.colorbar(contour, label='Error Ozone Value')
            plt.xticks(heatmap_data.columns[::5])
            plt.yticks(heatmap_data.index[::5])
            plt.xlabel('Longitude')
            plt.ylabel('Latitude')
            plt.title(f'Error Analysis for Ozone Levels over Delhi - {date_only}')
            plt.grid()
            plt.savefig(f'/home/sysadm/Downloads/delhi/delhi_o3/error_plot/Error Analysis for Ozone_plot_heat_{date_only}_delhi.png')
            plt.show()
```
