```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import geopandas as gpd

# Load the data for kriging and LSTM predictions for SO2
krig_so2 = pd.read_csv("/home/sysadm/Downloads/delhi/Delhi_2023_sep_So2_final.csv")
lstm_so2 = pd.read_csv("/home/sysadm/Downloads/delhi/delhi_so2/relu_predicted_6_30_so2.csv")

# Preprocess the data
krig_so2 = krig_so2.drop(columns=["latitude", "longitude", "2023-09-01", "2023-09-02", "2023-09-03", "2023-09-04", "2023-09-05"])
lstm_so2.index = lstm_so2["date"]
lstm_so2 = lstm_so2.drop(columns=['date'])

# Transpose LSTM predictions
relu_predicted_so2 = lstm_so2.transpose()

# Subtract kriging from LSTM and calculate the absolute values
result_df_numpy_so2 = np.abs(krig_so2.to_numpy() - relu_predicted_so2.to_numpy())

# Create a DataFrame with the absolute values
abs_array_df_so2 = pd.DataFrame(result_df_numpy_so2, columns=relu_predicted_so2.columns) 
krig_so2 = pd.read_csv("/home/sysadm/Downloads/delhi/Delhi_2023_sep_So2_final.csv")
abs_array_df_so2["longitude"] = krig_so2["longitude"]
abs_array_df_so2["latitude"] = krig_so2["latitude"]
result_df_so2 = abs_array_df_so2

# Create a GeoDataFrame from the result DataFrame
gdf_so2 = gpd.GeoDataFrame(result_df_so2, geometry=gpd.points_from_xy(result_df_so2['longitude'], result_df_so2['latitude']))

# Load the basemap shapefile
shapefile_path_so2 = './Delhi_shape_file/Delhi.shp'
basemap_so2 = gpd.read_file(shapefile_path_so2)

# Check if points are within the shapefile boundary
points_within_shapefile_so2 = gdf_so2[gdf_so2.geometry.within(basemap_so2.unary_union)]
gdf_within_shapefile_so2 = points_within_shapefile_so2.reset_index(drop=True)

# Iterate through each date column in the DataFrame
for column in gdf_within_shapefile_so2.columns[0:-3]:  # Exclude latitude and longitude columns
    # Check if the entire column has NaN values
    if not gdf_within_shapefile_so2[column].isnull().all():
        # Extract only the date part from the datetime string
        date_only_so2 = pd.to_datetime(column).date()

        # Pivot the DataFrame for the specific date
        heatmap_data_so2 = gdf_within_shapefile_so2.pivot(index='latitude', columns='longitude', values=column)

        # Create a contour plot if there are no NaN values for the specific date
        if not heatmap_data_so2.isnull().values.all():
            plt.figure(figsize=(10, 6))
            contour_so2 = plt.contourf(heatmap_data_so2.columns, heatmap_data_so2.index, heatmap_data_so2.values, cmap='winter')
            plt.colorbar(contour_so2, label='Error SO2 Value')
            plt.xticks(heatmap_data_so2.columns[::5])
            plt.yticks(heatmap_data_so2.index[::5])
            plt.xlabel('Longitude')
            plt.ylabel('Latitude')
            plt.title(f'Error Analysis for SO2 Levels over Delhi - {date_only_so2}')
            plt.grid()
            plt.savefig(f'/home/sysadm/Downloads/delhi/delhi_so2/error_plot/Error Analysis for SO2_plot_heat_{date_only_so2}_delhi.png')
            plt.show()
```
